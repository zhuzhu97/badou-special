```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def BGR2GRAY(img):
    Gray = np.zeros_like(img[:,:,0])
    for m,i in enumerate(img,0):
        for n,j in enumerate(i,0):
            b,g,r = j
            Gray[m][n] = .3*r+.59*g+.11*b 
    return Gray
img = cv2.imread('xiugou.jpg')

cv2.imshow('',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

### BGR2GRAY 

```python
Gray_code = BGR2GRAY(img)
cv2.imshow('',Gray_code)
cv2.waitKey()
cv2.destroyAllWindows()
```

### cv2.BGR2GRAY

```python
Gray_cv = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow('BGR2GRAY',Gray_cv)
cv2.waitKey()
cv2.destroyAllWindows()
```

### 

```python
threshold = 0.5
Nor_Gray = Gray_cv/np.max(Gray_cv)
pic = np.zeros_like(Gray_cv)
for m,i in enumerate(Nor_Gray,0):
    for n,j in enumerate(i,0):
        if Nor_Gray[m][n]>0.5: pic[m][n] = 1

cv2.imshow('dog',pic)
cv2.waitKey()
cv2.destroyAllWindows()
```
