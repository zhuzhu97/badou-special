```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
# Canny
def CannyThreshold(lowThreshold):  
    detected_edges = cv2.GaussianBlur(img,(3,3),0) # 
    detected_edges = cv2.Canny(detected_edges,
            lowThreshold,
            lowThreshold*ratio,
            apertureSize = kernel_size)  #
  
    re = cv2.getTrackbarPos('Canny', 'demo')
    dst = cv2.bitwise_and(img,img,mask = detected_edges)  #
    if re % 20 == 0:
        b,g,r = cv2.split(dst)          #BGR
        img_new1 = cv2.merge([r,g,b])   #RGBplt.imshow
        # 
        plt.subplot(2,3, (int)(re/20+1)), plt.xticks([]), plt.yticks([]), plt.title(re), plt.imshow(img_new1)
    cv2.imshow('demo',dst)
    #plt.imshow(dst)
    
lowThreshold = 0  
max_lowThreshold = 100  
ratio = 3  
kernel_size = 3  
  
img = cv2.imread('1.jpg')  
#gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)  #
#cv2.imshow("yuan",gray)
#cv2.waitKey()
cv2.namedWindow('demo')  
  
#,
cv2.createTrackbar('Canny','demo',lowThreshold, max_lowThreshold, CannyThreshold)  

CannyThreshold(0)  #  
    
if cv2.waitKey(0) == 27:  #wait for ESC key to exit cv2
        cv2.destroyAllWindows() 

 
    

```
